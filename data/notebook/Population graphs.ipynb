```python
labels = ['1996', '2001', '2006', '2013', '2018']
sizes = regional[['erp96', 'erp01', 'erp06', 'erp13', 'erp18']]

for i in regional.index:
    fig, ax = plt.subplots(figsize= (7,4))
    ax.bar(labels, sizes.iloc[i, :], width=0.5, color ='maroon', edgecolor='orange');
    plt.ylim(0, sizes.iloc[i, :].max() * 1.5)
    plt.xlabel("Years")
    plt.ylabel("Population counts");
    plt.title("Population counts across the years for " + regional['REGC2020_1'].iloc[i]);
    plt.savefig(r"C:\Users\asafi\Desktop\studies\BGU - Introduction to Web Mapping\project2\data\notebook\images\\" + regional['REGC2020_1'].iloc[i] + '.png');
    plt.show();
```
